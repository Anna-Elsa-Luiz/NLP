```python
with open('technologie_4.txt', 'r') as file:
    text = file.read()

print(text)
```

```python
import nltk
from nltk.stem import PorterStemmer #for stemming
from nltk.corpus import stopwords #for stopwards 
from nltk.stem import WordNetLemmatizer #for Lemmatization
import re #regrex
from sklearn.feature_extraction.text import  CountVectorizer #for BOW
from sklearn.feature_extraction.text import TfidfVectorizer  #for Tfidf
```

# 1. Tokenization

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
sentences = nltk.sent_tokenize(text)
sentences
```

Clean the paragraph 

```python
len(sentences)
```

```python
import re 
corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z0-9\s$]',' ',sentences[i]) # remove non alphabetic characters
    review= review.lower() #making the words lowercase
    review = review.split() 
    review =' '.join(review)
    corpus.append(review)
```

```python
sentences
```

# 2. Stemming

```python
stemmer = PorterStemmer()
```

```python
for i in corpus:
    words = nltk.word_tokenize(i) # it make sure that from the sentence I am getting the words 
    for word in words :
        if word not in set(stopwords.words('english')):
           print(stemmer.stem(word))
```

# 3.Lemmatization

```python
lemmatizer = WordNetLemmatizer()

for i in corpus:
    words = nltk.word_tokenize(i) # it make sure that from the sentence I am getting the words 
    for word in words :
        if word not in set(stopwords.words('english')):
           print(lemmatizer.lemmatize(word))
```

# 3. Bag Of Words 

```python
cv= CountVectorizer(binary=True,ngram_range=(1,3)) #ngrams :trigrams
```

```python
X = cv.fit_transform(corpus)
```
